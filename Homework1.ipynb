```python
#importing all the important libraries needed.
import pandas as pd
import numpy as np
import pylab as pl
from pandas import Series, DataFrame
```

```python
#Reading a csv file
loadData = pd.read_csv('http://www3.cs.stonybrook.edu/~skiena/591/hw1/country-data.csv')
loadData
```

```python
#To know the data type of the headers
loadData.dtypes
#print np.array(zip(list(loadData.columns), loadData.ix[0,:].map(lambda x: type(x))))
```

```python
# Missing values imputation

#removing the % from a value in column 'Milatary':
mil_index = 0
for b in loadData['Military Expenditure/GDP'].values:
    b = str(b)
    b = b.replace("%", "")
    loadData['Military Expenditure/GDP'][mil_index] = b
    mil_index +=1
    
loadData = loadData.replace('unknown',np.nan) 
loadData = loadData.replace('none',np.nan)    
    
#Typecasting strings to floats 
loadData['Literacy Rate (%)'] =  loadData['Literacy Rate (%)'].astype(float) 
loadData['Health Expenditure/GDP'] = map(float, loadData['Health Expenditure/GDP'])
loadData['Military Expenditure/GDP'] = map(float, loadData['Military Expenditure/GDP'])
loadData['Renewable Water Resources (cu km)'] = map(float, loadData['Renewable Water Resources (cu km)'])
loadData['Net Migration Rate/1000 Population'] = map(float, loadData['Net Migration Rate/1000 Population'])
loadData['Labor Force'] = map(float, loadData['Labor Force'])
loadData['Internet Users (circa 2009)'] = map(float, loadData['Internet Users (circa 2009)'])
loadData['Commercial Prime Lending Rate'] = map(float, loadData['Commercial Prime Lending Rate']) 
   

#filling missing values with the mean()
loadData = loadData.fillna(loadData.mean())

#print loadData['Military Expenditure/GDP'][130:150]
   
```

```python
import pandas as pd
import numpy as np
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
%matplotlib inline

#Convertiing latitude and logitude to plot on the graph
def conversion(old):
    direction = {'N':-1, 'S':1, 'E': -1, 'W':1}
    new = old.split()
    new_dir = new.pop()
    new.extend([0,0,0])
    return ((int(new[0])+int(new[1])/60.0+int(new[2])/3600.0) * (direction[new_dir]))

#source: stackoverflow
 
def get_marker_color(labor_force):
    
    if labor_force < 50000:
        return ('go')
    elif labor_force < 100000:
        return ('yo')
    elif labor_force < 10000000:
        return ('bo')
    else:
        return ('ro')
    
    
map = Basemap(projection='robin', lat_0=0, lon_0=0,resolution='c', area_thresh=1000.0)
 
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='green',lake_color='aqua')
map.drawcoastlines()
map.drawmapboundary()
 
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

#Read latitides, Longitudes and Internet Users
lats = loadData['Latitude of Capital']
lons = loadData['Longitude of Capital']
labor_force = loadData['Labor Force']


#Adding markers to the map
min_marker_size = 8
for lon, lat, labor in zip(lons, lats, labor_force):
    x,y = map(conversion(str(lon)), conversion(str(lat)))
    marker_color = get_marker_color(labor)
    map.plot(x, y, marker_color, markersize=min_marker_size)
    
#figsize(100, 100)
plt.show()

#source:http://matplotlib.org/basemap/users/geography.html
```

```python
y = loadData['GDP (PPP) in US $'].astype(float)
x1 = loadData['Health Expenditure/GDP'].astype(float)
x2 = loadData['Military Expenditure/GDP'].astype(float)
print y.shape, x1.shape, x2.shape

#shuffle
from sklearn.utils import shuffle
x1, y = shuffle(x1, y, random_state=1)
x2, y = shuffle(x2, y, random_state=1)

# create train_set and test_sets
train_size = 150
x1train = x1[:train_size]  # 150 rows for train
x1test = x1[train_size:]   # rest for test
x2train = x2[:train_size]  # selects first 250 rows (examples) for train set
x2test = x2[train_size:]   # selects from row 250 until the last one for test set
ytrain = y[:train_size]
ytest = y[train_size:]

from sklearn import linear_model
regress = linear_model.LinearRegression()

plt.scatter(x1train, ytrain)
plt.scatter(x1test, ytest)
plt.xlabel('Health Expenditure')
plt.ylabel('GDP')

x1_fit_train = x1train.reshape((x1train.shape[0],-1))
x1_fit_test = x1test.reshape((x1test.shape[0],-1))
y_fit_train = ytrain.reshape((ytrain.shape[0],-1))
y_fit_test = ytest.reshape((ytest.shape[0], -1))
regress.fit(x1_fit_train, y_fit_train)
import numpy as np
print("Health Expenditure/ GDP Training error: ", np.mean((regress.predict(x1_fit_train) - y_fit_train) ** 2))
print("Health Expenditure/ GDP Test error: ", np.mean((regress.predict(x1_fit_test) - y_fit_test) ** 2))

plt.scatter(x1_fit_train, y_fit_train,  color=np.random.rand(len(x1_fit_train)),s=20, c='b')

# Linear model plots
plt.plot(x1_fit_train, regress.predict(x1_fit_train), color='red', linewidth=5);
plt.xlabel('Health linear model (train)')
plt.ylabel('GDP')
plt.scatter(x1_fit_test, y_fit_test, color=np.random.rand(len(x1_fit_train)),s=20, c='b')

# Linear model plots
plt.plot(x1_fit_test, regress.predict(x1_fit_test), color='red', linewidth=5);
plt.xlabel('Health linear model (test)')
plt.ylabel('GDP')
```

```python
import numpy as np
import math
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from math import sin, cos, sqrt, atan2, radians

correlate = [loadData['Life Expectancy'],
             loadData['GDP (PPP) in US $'],
             loadData['Area (sq km)'],
             loadData['Land Boundaries (km)'],
             loadData['Coastline (km)'],
             loadData['Literacy Rate (%)'],
             loadData['Health Expenditure/GDP'],
             loadData['Military Expenditure/GDP'],
             loadData['Renewable Water Resources (cu km)'],
             loadData['Net Migration Rate/1000 Population'],
             loadData['Population'],
             loadData['Labor Force'],
             loadData['Internet Users (circa 2009)'],
             loadData['Commercial Prime Lending Rate']
             ]

col_index = 0    
while (col_index < len(correlate)-1):
    next_index = col_index + 1
    while (next_index < len(correlate)):
        coefficient = [correlate[col_index].corr(correlate[next_index])]
        #print col_index, next_index, coefficient
        if coefficient > 0.7:
            #print col_index,loadData['Country Name'][col_index]
            similar = loadData['Country Name'][col_index]
            print 'similar countries:', similar
        elif coefficient < 0:
            diff = loadData['Country Name'][col_index]
            print 'different countries:', diff
        next_index +=1
    col_index += 1 



#distance calculations
def findDistance(lat1,lon1,lat2,lon2,Rad=6373.0):
    diff_lat = lat2 - lat1
    diff_lon = lon2 - lon1
    var1 = (sin(diff_lat/2))**2 + cos(lat1) * cos(lat2) * (sin(diff_lon/2))**2
    var2 = 2 * atan2(sqrt(var1), sqrt(1-var1))
    distance = Rad * var2
    return distance

#source:http://stackoverflow.com/questions/15736995/how-can-i-quickly-estimate-the-distance-between-two-latitude-longitude-points

newlist = [ \
                 'Country Name', \
                 'Latitude of Capital', \
                 'Longitude of Capital'\
                ]

Euclidian_Distance = pd.DataFrame(data=None, columns=['Country1','Country2','Distance'])

for i in np.arange(50):
    src_country = loadData.loc[i,newlist[0]]
    src_lat = loadData.loc[i,newlist[1]]
    src_lon = loadData.loc[i,newlist[2]]
    #print(i)
    for j in np.arange(50):
        dest_country = loadData.loc[j,newlist[0]]
        dest_lat = loadData.loc[j,newlist[1]]
        dest_lon = loadData.loc[j,newlist[2]]
        lat1 = math.radians(conversion(str(src_lat)))
        lon1 = math.radians(conversion(str(src_lon)))
        lat2 = math.radians(conversion(str(dest_lat)))
        lon2 = math.radians(conversion(str(dest_lon)))
        temp = np.asarray([src_country, dest_country,float(findDistance(lat1,lon1,lat2,lon2))])
        Euclidian_Distance.loc[len(Euclidian_Distance)+1] = temp
        #print(src_country, dest_country,findDistance(lat1,lon1,lat2,lon2))
Euclidian_Distance = Euclidian_Distance[Euclidian_Distance['Country1']!=Euclidian_Distance['Country2']].sort(columns='Distance')
print ("Distance Table")
Euclidian_Distance
```
